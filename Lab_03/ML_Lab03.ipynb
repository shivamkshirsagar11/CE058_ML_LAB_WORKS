```python
import pandas as pd
from sklearn import preprocessing as pre
from sklearn.naive_bayes import GaussianNB as gnb, MultinomialNB as mnb
```

# **1st Example**

---



# With **Gaussian**

```python
if __name__ == "__main__":
  label = pre.LabelEncoder()
  data = pd.read_csv("/content/drive/MyDrive/SEM-6/ML/Data_Lab03/Dataset1.csv")
  # print(data)
  X = data.iloc[:,0:-1].values
  Y = data.iloc[:,-1:].values
  # print(Y)
  for i in range(X.shape[1]):
    X[:,i] = label.fit_transform(X[:,i])
    X[:,i] = X[:,i].astype(int)
  Y[:,0] = label.fit_transform(Y[:,0])
  Y = Y[:,0].astype(int)
  # print(Y[:,0])
  # print("features...")
  print(X)
  # print("Class prior")
  # print(Y)
  features = tuple(zip(X[:,0],X[:,1],X[:,2],X[:,3]))
  Model = mnb()
  Model.fit(features,Y)
  print(f'Pridicted: {Model.predict([[2,0,2,0]])}')
```

# With **Multinomial**

```python
if __name__ == "__main__":
  data = pd.read_csv("/content/drive/MyDrive/SEM-6/ML/Data_Lab03/Dataset1.csv")
  X = data.iloc[:,:-1].values
  Y = data['Play'].values
  # print(Y)
  label = pre.LabelEncoder()
  for i in range(X.shape[1]):
    X[:,i] = label.fit_transform(X[:,i])
    X[:,i] = X[:,i].astype(int)
  Y = label.fit_transform(Y)
  features = tuple(zip(*[X[:,i] for i in range(X.shape[1]) ]))
  
  Model = gnb()
  # Model.var_smoothing = 10
  Model.get_params(deep=True)
  Model.fit(features,Y)
  print(f'Pridicted: {Model.predict([[2,0,2,0]])}')
  print(f'Pridicted Log joint prob: {Model.predict_log_proba([[2,0,2,0]])}')
  print(f'Pridicted joint prob: {Model.predict_proba([[2,0,2,0]])}')
  print(f'Smoothing param: {Model.var_smoothing}')
```
